```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

#ax.legend(loc='lower left')
#ax.legend(loc=4)
ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e8d420',label='x vs x',lw=3,ls='--') #RGB HEX Code

ax.plot(x,x+1,color='purple',label= 'x vs x+1',linewidth=5,linestyle='-.')


# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])


lines = ax.plot(x,x+1,color='purple',label= 'x vs x+1',linewidth=5)
lines[0].set_dashes([1,2,1,2,10,2])

# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='purple',label= 'x vs x+1',linewidth=2,marker='+',ls='--',ms=10)

ax.plot(x,x+1,color='red',label= 'x vs x+1',linewidth=2,marker='s',ls='--',markersize=10, 
        markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange')

ax.legend()
```
