```python
fig = plt.figure()

# Large Axes

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# Small Axes

axes2 = fig.add_axes([1,1,2,2])

axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')


plt.show()
```
